```python
from ROOT import TH1F, TGraph, TCanvas
import lhapdf
lhapdf.availablePDFSets()
```

```python
#p = lhapdf.mkPDF("CT10nlo")
p = lhapdf.mkPDF("NNPDF31_nnlo_as_0118")
```

```python
h_fuv = TH1F("fuv","valence up quark",100,0,1)
h_fdv = TH1F("fdv","valence down quark",100,0,1)
h_fuvx = TH1F("fuvx","valence up quark fx",100,0,1)
h_fdvx = TH1F("fdvx","valence down quark fx",100,0,1)
h_fuvx.SetMaximum(1)
h_fdvx.SetMaximum(1)
h_fuvx.SetStats(0)
h_fdvx.SetStats(0)
```

```python
Q = 5.0
x = 0.01
n = 0
```

```python
for i in range(100):
  x = h_fuv.GetBinCenter(i+1)
  h_fdv.SetBinContent(i+1,(p.xfxQ(1,x,Q)-p.xfxQ(-1,x,Q))/x)
  h_fdvx.SetBinContent(i+1,p.xfxQ(1,x,Q)-p.xfxQ(-1,x,Q))
  h_fuv.SetBinContent(i+1,(p.xfxQ(2,x,Q)-p.xfxQ(-2,x,Q))/x)
  h_fuvx.SetBinContent(i+1,p.xfxQ(2,x,Q)-p.xfxQ(-2,x,Q))
```

```python
c1 = TCanvas("c1","LHAPDF Example",0,0,700,500)
c1.Divide(2,1)
c1.cd(1)
h_fdvx.Draw("c")
h_fdvx.GetXaxis().SetTitle("x")
h_fdvx.GetYaxis().SetTitle("x*f_{d}(x)")
print("Integral of valence down quark pdf = ", h_fdv.Integral("width"))
```

```python
c1.cd(2)
h_fuvx.Draw("c")
h_fuvx.GetXaxis().SetTitle("x")
h_fuvx.GetYaxis().SetTitle("x*f_{u}(x)")
print("Integral of valence up quark pdf = ",h_fuv.Integral("width"))
c1.Update()
c1.Draw()
c1.Print("lhapdf_example.pdf")
```
